```python
# I have fitted a binary logistic regression model using AGE, HIV exposure, malnutrition and HIV status as covariates
#although not significant,HIV status makes HIV exposure significantinthe model! Antony can give his input!
# without HIV status in the model, HIV exposure is not significant at P<0.05
import statsmodels.formula.api as smf
model = smf.logit("outcome ~ AGEINMONTHS + C(HIV_exposed)+ C(Malnutrition) + C(HIV_Infected)", data = df).fit()
model.summary()
```

```python
#The above results are the log odds which is not intuitive (straight forward) in it's interpretation.
#Let's convert this to odds ratio and interpret the model again.
#To convert the log odds coefficients and confidence intervals, one needs to take the exponential of the values.

model_odds = pd.DataFrame(np.exp(model.params), columns= ['OR'])
model_odds['z-value']= model.pvalues
model_odds[['2.5%', '97.5%']] = np.exp(model.conf_int())

model_odds
```

```python
#INTERPRETATION
#The overall model indicates the model is better than using the mean to predict death from pneumonia, F(4, 141) < 0.0000.
#Converting to odd ratios (OR) is much more intuitive in the interpretation. Where,
#OR = 1, same odds
#OR < 1, fewer/decrease in odds
#OR > 1, greater/increase in odds
#The odds of dying increases by a factor of 4.7 if you are HIV exposed; 
#likewise, the odds of dying increases by a factor of 4.9 if you are malnourished.
#HIV infection is not significant but it somehow makes HIV exposure significant
#The odds of dying decreases by a factor of 0.8 for every unit increase in a child'sage in months.
```
